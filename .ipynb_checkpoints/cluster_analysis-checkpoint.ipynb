```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_df = pd.read_csv("mall_customers.csv")
```

```python
assert customer_df.duplicated().any() == False
```

```python
for index, row in customer_df.iterrows():
    if row.isna().any():
        print(f"{index} missing values exist")
```

```python
cdf_by_age = customer_df.copy()

print(customer_df["Age"].unique())

print(customer_df["Gender"].unique())

print(customer_df.dtypes)

customer_df
```

Already, there are few problems with our table. 

First, "CustomerID" should be the table index, "Annual Income" should be fully expressed as it would be more sensible visually. Also, there should be a category or range for the "Age" field and finally, the table column names should have a simpler casing and not contain special characters for the sake of indexing.

```python
labels = ["10 - 19", "20 - 29", "30 - 39", "40 - 49", "50 - 59", "60 - 69", "70+"]

ranges = [10, 19, 29, 39, 49, 59, 69, np.inf]

gender_dict = {"Male": 1, "Female": 2}

cdf_by_age["gender_class"] = cdf_by_age["Gender"].replace(gender_dict)

cdf_by_age["Age_group"] = pd.cut(customer_df["Age"], bins = ranges, labels = labels)

cdf_by_age["Annual Income (k$)"] = cdf_by_age["Annual Income (k$)"] * 1000

cdf_by_age["total_amount"] = (cdf_by_age["Annual Income (k$)"]) * (cdf_by_age["Spending Score (1-100)"] / 100)

new_order = ['CustomerID', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'total_amount', 'gender_class', 'Gender', 'Age_group']

cdf_by_age = cdf_by_age[new_order]

new_column_names = {'CustomerID': 'customer_id', 'Age': 'age', 'Annual Income (k$)': 'annual_income', 'Spending Score (1-100)': 'spending_score', 'Gender': 'gender', 'Age_group': 'age_group'}

cdf_by_age = cdf_by_age.rename(columns = new_column_names)

cdf_by_age
```

```python
cdf_by_age.describe()
```

```python
plt.figure(dpi = 100)
sns.scatterplot(
    data = cdf_by_age, x = "customer_id", y = "spending_score", hue = "gender",
    size = "age", sizes=(20, 200), alpha = 0.5
)

plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1.02))

plt.show()
```

```python
# Plot of customer id vs spending score

plt.figure(dpi = 100)

x = cdf_by_age["customer_id"]
y = cdf_by_age["age"]

sns.scatterplot(
    data = cdf_by_age, x = x, y = y, hue = "age_group",
    sizes=(20, 200), alpha = 0.7
)

plt.xlabel("Customer ID")
plt.ylabel("Age")

plt.ylim(0, 80)

plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1.02))

plt.title("Plot of customers by their Age")

plt.show()
```

```python
# Plot of customer id vs spending score

plt.figure(dpi = 100)

x = cdf_by_age["age"]
y = cdf_by_age["spending_score"]

sns.scatterplot(
    data = cdf_by_age, x = x, y = y, hue = "age_group",
    sizes=(20, 200), alpha = 0.7
)

plt.xlabel("Age")
plt.ylabel("Spending Score")

plt.ylim(0, 80)

plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1.02))

plt.title("Plot of Age by their Spending Score")

plt.show()
```

```python
# Histogram plot of customers by their age groups 

groups_label = cdf_by_age["age_group"].unique()

print(groups_label)

age_sorted = cdf_by_age.sort_values(by='age_group', ascending=True)

plt.figure(dpi = 100)

plt.hist(age_sorted["age_group"], bins = len(labels), edgecolor = "white")

plt.xlabel("Age Intervals")
plt.ylabel("Frequency")

plt.title("Distribution of Age Groups")

plt.xticks(rotation = 45)

plt.show()
```

```python
# Histogram plot of customers by their gender 

plt.figure(figsize = (10, 7))

gender = cdf_by_age["gender"].unique()
count = [len(cdf_by_age[cdf_by_age["gender"] == "Male"]), len(cdf_by_age[cdf_by_age["gender"] == "Female"])]
bar_labels = cdf_by_age["gender"].unique()
bar_colors = ['tab:red', 'tab:blue']

plt.bar(gender, count, label = bar_labels, color = bar_colors)
# print(len(cust_by_age[cust_by_age["Gender"] == "Male"]))
# plt.bar(age_sorted["Gender"], height = 10, bins = len(labels), edgecolor = "white")

plt.xlabel("Gender")
plt.ylabel("Frequency")

plt.title("Distribution by Gender")

plt.xticks(rotation = 45)
plt.yticks(np.linspace(0, 120, 13))

plt.show()
```

```python
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.scatter(cdf_by_age["age"], cdf_by_age["annual_income"], cdf_by_age["spending_score"], marker = "o")

ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Spending Score')

plt.show()
```

```python
# Correlation matrix and confusion matrix

figures = cdf_by_age.iloc[:, 1:5]
gen_fig = cdf_by_age.iloc[:, [1, 2, 3, 4, 6]]
corr_matrix = figures.corr(method = "pearson")
corr_matrix
```

```python
sns.heatmap(corr_matrix, cmap=sns.cubehelix_palette(as_cmap = True), vmin = -0.4, vmax = 1, annot = True, fmt=".3f")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Pairplot
g = sns.PairGrid(gen_fig, hue = 'gender')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
plt.show()
```
